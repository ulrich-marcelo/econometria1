```python
import yfinance as yf
# Define el símbolo (ticker) y las fechas de inicio y fin
ticker = 'appl'
start_date = '2014-01-01'
end_date = '2015-12-31'
# Descarga los datos utilizando yfinance
F_data = yf.download(ticker, start=start_date, end=end_date)
# Imprime los primeros registros de los datos descargados
print(f'F_data.head(): \n{F_data.head()}\n')
print(f'F_data.tail(): \n{F_data.tail()}\n')
```
